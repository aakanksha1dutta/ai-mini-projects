```python
import numpy as np
from math import log
```

```python
def parse_libsvm_file(filename):
    labels = []
    features = []

    with open(filename, 'r') as file:
        for line in file:
            elements = line.strip().split(' ')
            labels.append(int(elements[0]))
            feature_vector = {}
            for item in elements[1:]:
                index, value = item.split(':')
                feature_vector[int(index)] = float(value)
            features.append(feature_vector)

    return labels, features
```

```python
labels, features = parse_libsvm_file("./a7a.train")
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def initialize_params(num_features):
    # Initialize weights with zeros 
    w = np.zeros(num_features)
    w0 = 0
    return w0, w

def logistic_reg(X, y, learning_rate, N):
    num_samples, num_features = X.shape
    b, w = initialize_params(num_features)

    for i in range(N):
        # compute predicted
        dot_prod = np.dot(X, w)
        z =  dot_prod + b
        A = sigmoid(y*z)

        # Compute loss
        loss = np.sum(-1*np.log(A))
        
        # Backward pass: compute gradients
        dw = (1-A)*y*X.T
        db = np.mean(A)

        # Update parameters
        w -= learning_rate * dw
        b -= learning_rate * db

        # Print the loss every 100 iterations
        if i%100==0:
            print (f"Iterations {i}: Loss {loss}")

    return w, b

```
